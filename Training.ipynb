```python
import os
import shutil

# Define source and destination directories
source_train_dir = '/content/PlantDoc-Object-Detection-Dataset/TRAIN'
source_test_dir = '/content/PlantDoc-Object-Detection-Dataset/TEST'
destination_train_images_dir = '/content/PlantDoc-Object-Detection-Dataset/train_images_only'
destination_test_images_dir = '/content/PlantDoc-Object-Detection-Dataset/test_images_only'

# Create destination directories if they don't exist
os.makedirs(destination_train_images_dir, exist_ok=True)
os.makedirs(destination_test_images_dir, exist_ok=True)

def copy_images_only(source_dir, destination_dir):
    for root, _, files in os.walk(source_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                source_file = os.path.join(root, file)
                destination_file = os.path.join(destination_dir, file)
                try:
                    shutil.copy(source_file, destination_file)
                except Exception as e:
                    print(f"Error copying {file}: {e}")

# Copy images from train folder
copy_images_only(source_train_dir, destination_train_images_dir)

# Copy images from test folder
copy_images_only(source_test_dir, destination_test_images_dir)

print("Image extraction completed.")
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('train_labels.csv')

# Define class labels
class_labels = {
    'Cherry leaf': 0,
    'Peach leaf': 1,
    'Corn leaf blight': 2,
    'Apple rust leaf': 3,
    'Potato leaf late blight': 4,
    'Strawberry leaf': 5,
    'Corn rust leaf': 6,
    'Tomato leaf late blight': 7,
    'Tomato mold leaf': 8,
    'Potato leaf early blight': 9,
    'Apple leaf': 10,
    'Tomato leaf yellow virus': 11,
    'Blueberry leaf': 12,
    'Tomato leaf mosaic virus': 13,
    'Raspberry leaf': 14,
    'Tomato leaf bacterial spot': 15,
    'Squash Powdery mildew leaf': 16,
    'grape leaf': 17,
    'Corn Gray leaf spot': 18,
    'Tomato Early blight leaf': 19,
    'Apple Scab Leaf': 20,
    'Tomato Septoria leaf spot': 21,
    'Tomato leaf': 22,
    'Soyabean leaf': 23,
    'Bell_pepper leaf spot': 24,
    'Bell_pepper leaf': 25,
    'grape leaf black rot': 26,
    'Potato leaf': 27,
    'Tomato two spotted spider mites leaf': 28
}

output_dir = '/content/PlantDoc-Object-Detection-Dataset/train/labels/'
os.makedirs(output_dir, exist_ok=True)

# Function to normalize coordinates
def normalize_coordinates(row):
    if row['width'] == 0 or row['height'] == 0:
        row['width'] = 50
        row['height'] = 100

    x_center = (row['xmin'] + row['xmax']) / 2 / row['width']
    y_center = (row['ymin'] + row['ymax']) / 2 / row['height']
    width = (row['xmax'] - row['xmin']) / row['width']
    height = (row['ymax'] - row['ymin']) / row['height']
    return x_center, y_center, width, height

# Convert annotations
for filename, group in df.groupby('filename'):
    txt_filename = os.path.join(output_dir, os.path.splitext(filename)[0] + '.txt')
    with open(txt_filename, 'w') as f:
        for _, row in group.iterrows():
            class_id = class_labels[row['class']]
            normalized_coords = normalize_coordinates(row)
            if normalized_coords is not None:
                x_center, y_center, width, height = normalized_coords
                f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

df = pd.read_csv('test_labels.csv')
output_dir = '/content/PlantDoc-Object-Detection-Dataset/test/labels/'
os.makedirs(output_dir, exist_ok=True)

# Convert annotations
for filename, group in df.groupby('filename'):
    txt_filename = os.path.join(output_dir, os.path.splitext(filename)[0] + '.txt')
    with open(txt_filename, 'w') as f:
        for _, row in group.iterrows():
            class_id = class_labels[row['class']]
            normalized_coords = normalize_coordinates(row)
            if normalized_coords is not None:
                x_center, y_center, width, height = normalized_coords
                f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

print("Conversion to YOLOv8 format completed.")
```
